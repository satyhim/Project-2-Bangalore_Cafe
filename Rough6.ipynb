#Dealing with Null values

Now We can see we Null values for total five columns , if we drop all the null values we will lost huge data from our dataframe. 

So to avoid data loss we are going to replace null values with some values assigned at each place of null locations.

We can do it will below process:

```python
location_groupby=df.groupby('listed_in(city)')
location_groupby.describe()
```

We can see from above table data, we have 30 listed city locations from which Rajajinagar has smallest count , so we are replacing all null values with Rajajinagar in listed in city column as :

```python
df['listed_in(city)'] = df['listed_in(city)'].fillna('Unknown')
null=df.isna().sum()
null
```

So we replaced all null values with "**Unknown**" in liste city column, Now we are going to replace null values from dish_liked by giving some dish names into null values:

```python
df['dish_liked'] = df['dish_liked'].fillna('pizza_pasta_coffee')
null=df.isna().sum()
null
```

Now to replace null values from approx_cost(for two people) we need to check all existing cost values and we can calculate its median value so that we can replace with that value as:

```python
df.head(2)
```

```python
#mean_value=df['approx_cost(for two people)'].mean(skipna = True)
#mean=approx_cost(for two people).mean(axis = 1, skipna = True)
#df['approx_cost(for two people)'].mean(skipna = True)
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost_Two_People'}, inplace = True)
df.head(2)
```

```python
# We are getting error while replacing null values with median of cost of two people because the numbers having commas.
# So we need to remove commas first as:
df['Cost_Two_People'] = df['Cost_Two_People'].str.replace(',','')
```

```python
#gfg['G2'].fillna(value=mean_value, inplace=True)
median_value=df['Cost_Two_People'].median()
df['Cost_Two_People'].fillna(value=median_value, inplace=True)
```

```python
null=df.isna().sum()
null
```

As can see we still have two columns which contacin null values those are rate and listed in (type).
Now we are going to replace listed in type null values to "Cafes", as : 

```python
df['listed_in(type)'] = df['listed_in(type)'].fillna('Cafes')
null=df.isna().sum()
null
```

As we know that rate column has rating out of 5 , so we need to remove /5 from all the rows so as to get single integer number, we can replace method as :

```python
df['rate'] = df['rate'].replace('NEW',np.NaN)
df['rate'] = df['rate'].replace('-',np.NaN)

df['rate'] = df.loc[:,'rate'].replace('[ ]','',regex = True)
df['rate'] = df['rate'].astype(str)
df['rate'] = df['rate'].apply(lambda r: r.replace('/5',''))
df['rate'] = df['rate'].apply(lambda r: float(r))
```
